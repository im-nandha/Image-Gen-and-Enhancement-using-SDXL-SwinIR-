```python
import torch
from diffusers import StableDiffusionXLPipeline

model_id = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = StableDiffusionXLPipeline.from_pretrained(model_id, torch_dtype=torch.float16, variant="fp16")
pipe.to("cuda")
```

```python

from IPython.display import display
from PIL import Image

prompt = input("Enter your prompt for image generation: ")

# Generate image
image = pipe(prompt).images[0]


display(image)

image.save("sdxl_output.png")
print("Image saved as sdxl_output.png")
```
